```python
import pandas as pd
```

```python
housing_user_df = pd.read_csv('../../data/raw/nyc_user_housing_violation.csv')
housing_gov_df = pd.read_csv('../../data/raw/nyc_gov_housing_violation.csv')
```

```python
set(housing_gov_df['currentstatus'])
```

```python
housing_user_df['buildingid'] = housing_user_df['buildingid'].astype(str)
housing_gov_df['buildingid'] = housing_gov_df['buildingid'].astype(str)

housing_user_df['apartment'] = housing_user_df['apartment'].astype(str)
housing_gov_df['apartment'] = housing_gov_df['apartment'].astype(str)

housing_user_df['location'] = housing_user_df['buildingid'] + housing_user_df['apartment']
housing_gov_df['location'] = housing_gov_df['buildingid'] + housing_gov_df['apartment']
```

```python
housing_user_df['statusdate'] = pd.to_datetime(housing_user_df['statusdate']).dt.normalize()
housing_gov_df['currentstatusdate'] = pd.to_datetime(housing_gov_df['currentstatusdate']).dt.normalize()
```

```python
merged_df = pd.merge(
    housing_user_df,
    housing_gov_df,
    how="inner",
    left_on=["location", "statusdate"],
    right_on=["location", "currentstatusdate"],
)
```

```python
merged_df.to_csv('../../data/processed/aligned_housing_code.csv')
```
